```python
import requests
from bs4 import BeautifulSoup

search_word = ["陈息坤",]
page_index = None

for i in range(1, 51):
    url = "http://school.freekaoyan.com/sh/shu/daoshi/index"
    if i > 1:
        url += f"_{i}"
    url += ".shtml"

    print(url)

    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    print(soup.prettify())
    
    if search_word in str(soup):
        page_index = i
        break

print(f"The word '{search_word}' was found on page {page_index}" if page_index else "The word was not found in any of the pages.")
```
